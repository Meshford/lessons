- **p-value:** –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ —Å–æ–≤–ø–∞–¥–µ–Ω–∏—è. –ï—Å–ª–∏ p < 0.05 ‚Äî –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è.
- **–ü—Ä–∏–º–µ—Ä:**
  ```python
  from scipy.stats import ttest_ind
  
  group1 = np.random.normal(0, 1, 100)
  group2 = np.random.normal(1, 1, 100)
  t_stat, p_value = ttest_ind(group1, group2)
  print(f'p-value: {p_value:.3f}')
  ```
- **–ê–Ω–∞–ª–æ–≥–∏—è:** –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π —Ç–µ—Å—Ç ‚Äî –∫–∞–∫ –¥–µ—Ç–µ–∫—Ç–æ—Ä –ª–∂–∏: –æ–Ω –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –Ω–∞—Å–∫–æ–ª—å–∫–æ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–Ω–æ –≤–∞—à–µ —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ [[6]].

- –ü—Ä–∏–≤–µ–¥–∏—Ç–µ –ø—Ä–∏–º–µ—Ä—ã, –≥–¥–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ —Ç–µ—Å—Ç—ã –ø–æ–ª–µ–∑–Ω—ã (–Ω–∞–ø—Ä–∏–º–µ—Ä, —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π).
- –û–±—ä—è—Å–Ω–∏—Ç–µ, –ø–æ—á–µ–º—É –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –≤–∞–∂–Ω–∞ –¥–ª—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π.
